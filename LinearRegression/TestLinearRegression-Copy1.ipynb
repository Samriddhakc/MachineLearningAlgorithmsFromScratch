```python
import numpy as np 
from sklearn.datasets import make_regression
from matplotlib import pyplot
from mpl_toolkits.mplot3d import axes3d, Axes3D
# generate regression dataset
X, Z = make_regression(n_samples=100, n_features=2, noise=15)
#plot regression dataset
fig =pyplot.figure()
ax =pyplot.axes(projection='3d')
ax.scatter(np.squeeze(X[:,0]), np.squeeze(X[:,1]), Z, color='blue')
ax.set_title('wireframe');
y=y.reshape((y.shape[0],1))
# #X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
x_mean=np.mean(X,axis=0)
x_sd=np.std(X,axis=0)
X=(X-x_mean)/x_sd
Z_mean=np.mean(Z)
Z_sd=np.std(Z)
Z=(Z-Z_mean)/Z_sd
X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
#Regression using normal equation to find the exact solution 
linear_reg=LinearRegression(2)
params=linear_reg.fit_with_normal_equation(X_refined,Z)
#params=params.reshape((2,1))
```

```python
exact_hypothesis=np.dot(X_refined,params)
exact_hypothesis=exact_hypothesis.reshape((100,1))
#pyplot.scatter(X,y)
fig =pyplot.figure()
ax =pyplot.axes(projection='3d')
ax.scatter(X_refined[:,0],X_refined[:,1], Z, color='blue')
ax.plot(np.squeeze(X_refined[:,0]),np.squeeze(X_refined[:,1]),exact_hypothesis,color='red')
pyplot.show()
```

```python
#Regression using gradient descent to find a good enought solution using the grad descent algo 
params_2=linear_reg.fit_with_gradient_descent(X_refined,y,500,0.1)
```

```python
approx_hypothesis=np.dot(X_refined,params_2)
#pyplot.scatter(X,y)
pyplot.scatter(X,y)
pyplot.plot(X_refined[:,1],approx_hypothesis.squeeze(),'-r',color='black')
pyplot.show()
```
