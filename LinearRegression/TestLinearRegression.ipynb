```python
import numpy as np 
from sklearn.datasets import make_regression
from matplotlib import pyplot
# generate regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=15)
# plot regression dataset
pyplot.scatter(X,y)
#pyplot.show()
y=y.reshape((y.shape[0],1))
#X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
x_mean=np.mean(X)
x_sd=np.std(X)
X=(X-x_mean)/x_sd
X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
linear_reg=LinearRegression(1)
params=linear_reg.fit_with_gradient_descent(X_refined,y,150,0.005)
```

```python
hypothesis=np.dot(X_refined,params)
#pyplot.scatter(X,y)
pyplot.scatter(X,y)
pyplot.plot(X_refined[:,1],hypothesis.squeeze(),'-r')
pyplot.show()
```
