```python
from __future__ import division
import import_ipynb
from LinearRegression import LinearRegression
```

```python
import numpy as np 
from sklearn.datasets import make_regression
from matplotlib import pyplot
# generate regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=15)
# plot regression dataset
pyplot.scatter(X,y)
#pyplot.show()
y=y.reshape((y.shape[0],1))
#X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
x_mean=np.mean(X,axis=0)
x_sd=np.std(X,axis=0)
X=(X-x_mean)/x_sd
y_mean=np.mean(y)
y_sd=np.std(y)
y=(y-y_mean)/y_sd
X_refined=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
#Regression using normal equation to find the exact solution 
linear_reg=LinearRegression(1)
params=linear_reg.fit_with_normal_equation(X_refined,y)
#params=params.reshape((2,1))
```

```python
exact_hypothesis=np.dot(X_refined,params)
approx_hypothesis=np.dot(X_refined,params)
#pyplot.scatter(X,y)
pyplot.scatter(X,y)
pyplot.plot(X_refined[:,1],approx_hypothesis.squeeze(),'-r',color='black')
pyplot.show()
```

```python
#Regression using gradient descent to find a good enought solution using the grad descent algo 
params_2=linear_reg.fit_with_gradient_descent(X_refined,y,100,0.01)
```

```python
approx_hypothesis=np.dot(X_refined,params_2)
#pyplot.scatter(X,y)
pyplot.scatter(X,y)
pyplot.plot(X_refined[:,1],approx_hypothesis.squeeze(),'-r',color='red')
pyplot.show()
```
