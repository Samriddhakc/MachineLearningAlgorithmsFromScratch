```python
class LinearRegression:
    
    def __init__(self,data_dimension):
      
        self.theta=np.ones((data_dimension+1,1))


    def loss(self,X,y):
        
        """Example function with PEP 484 type annotations.

        Args:
            X: Matrix containing the features for every example.
            X=[x1,x2,x3,...].T (train_examples*features) 
            y: A column vector containing the expected output for all train examples. 

        Returns:
            loss for the linear regression which is computed as a mean squared loss. 
            ."""

        m=X.shape[0]
        J=(1/(2*m))*np.sum((np.dot(X,self.theta)-y)**2)
        
        return J 

    
    
    def fit_with_gradient_descent(self,X,y, num_iterations=100,learning_rate=0.01):
        

        """Example function with PEP 484 type annotations.

        Args:
            X: Matrix containing the features for every example.
            X= [x1,x2,x3,...].T (train_examples*features) 
            y: A column vector containing the expected output for all train examples. 
            num_interations: number of iterations to run the gradient descent algo. 
            learning_rate: a hyperparameter for the network

        Returns:
            returns the optimized parameter theta 
        """
        m=X.shape[0]
        #print("The shape of X is",X.shape)
        dtheta=(1/m)*np.dot((np.dot(X,self.theta)-y).T,X)
        dtheta=dtheta.T
        #print("The shape of dtheta is",dtheta.shape)
        
        for i in range(num_iterations):
            self.theta=self.theta-learning_rate*dtheta
            if (i%10==0):
                #print(self.theta.shape)
                print("The loss after"+str(i)+"iteration is"+str(self.loss(X,y)))
        return self.theta 


             
    def fit_with_normal_equation(self,X,y):

        """Example function with PEP 484 type annotations.

        Args:
            param1: The first parameter.
            param2: The second parameter.

        Returns:
            The return value. True for success, False otherwise.

        """

        combined_matrix=np.dot(X.T,X)
        combined_matrix=np.dot(np.linalg.pinv(combined_matrix),X.T)
        optimized_theta=np.dot(combined_matrix,y) 
        print("The dimension of optimized theta is",optimized_theta.shape)
        return optimized_theta 

    '''def accuracy():

               """Example function with PEP 484 type annotations.

        Args:
            param1: The first parameter.
            param2: The second parameter.

        Returns:
            The return value. True for success, False otherwise.

        '''
```
